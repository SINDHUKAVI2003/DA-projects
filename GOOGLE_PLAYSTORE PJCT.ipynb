# Q8 : Convert the column "Reviews'' to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach.

```python
# To check the count of values of 'Reviews'
b['Reviews'].value_counts()
```

```python
# To convert into numeric
b['Reviews'] = pd.to_numeric(b['Reviews'], errors='coerce')
b['Reviews']
```

```python
# To visualize the outliers using boxplot
plt.figure(figsize=(8,3))
sns.boxplot(x=b["Reviews"])
plt.show()
```

```python
# To treat the outliers using IQR 
Q1 = b['Reviews'].quantile(0.25)
Q3 = b['Reviews'].quantile(0.75)
IQR = Q3 - Q1

IQR_threshold = 1.5

lower = b['Reviews'] < (Q1 - IQR_threshold * IQR)
higher = b['Reviews'] > (Q3 + IQR_threshold * IQR)


b.loc[lower, 'Reviews'] = np.log1p(b.loc[lower, 'Reviews'])
b.loc[higher, 'Reviews'] = np.log1p(b.loc[higher, 'Reviews'])

outliers = b[lower | higher]

if outliers.empty:
    print('No outliers found')
else:
    print("Outliers found:",len(outliers))
```

```python
# To display the accurate value 
b['Reviews'] = np.log1p(b['Reviews']) 
b['Reviews'].head(10)
```

# INFERENCE

* The column 'Reviews' is converted into numerical data type.
* The log transformation is used.
